Discreet- How many people are in a class?

Uniform- Coin flip, all outcomes have the same probability.

Expected Value

Law of Large Numbers- As sample size increases the sample mean gets closer to the pop.  mean. 
Uniform Distribution- For a random variable, expected value is (a+b)/2, where a is max, and b is min. 
Binomial Distribution- Mean is the expected value, which is equal to n trials*p probability.
